```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0,stratify=y)
```

```python
X_train.shape
```

```python
def prediction(X,weights):
    y_est=np.dot(X,weights)
    return y_est
```

```python
def cost_function(y_est,y):
    m=y.shape[1]
    J=(1/(2*m))*np.sum(np.square(y_est-y))
    return J
```

```python
def gradient_descent(X,y,y_est):
    m=y.shape[1]
    dy=(y_est-y)
    dw=np.dot(X.T,dy)
    return dw
```

```python
def update_parameters(weights,dw,learning_rate):
    weights=weights-(learning_rate*dw)
    return weights
```

```python
def linear_regression(X,y,weights):
    y_est=prediction(X,weights)
    dw=gradient_descent(X,y,y_est)
    weights=update_parameters(weights,dw,0.0001)
    return weights
```

```python
for i in range(1000):
    weights=linear_regression(X_train,y_train,weights)
```
